```python
#create labels
a = [1,7,2]
myvar = pd.Series(a,index = ["x","y","z"])
print(myvar)
```

```python
#key/value object
calories = {"day1": 420,"day2":380,"day3":670}
myvar = pd.Series(calories)
print(myvar)
```

```python
#dataframes
import pandas as pd
data = {
    "calories":[420,380,390],
    "duration":[50,40,45]
}
myvar = pd.DataFrame(data)
print(myvar)
```

```python
#locate Row.py
import pandas as pd
data = {
    "calories":[420,380,390],
    "duration":[50,40,45]
}
myvar = pd.DataFrame(data)
print(myvar)
print(myvar.loc[[0,1]])
```
