```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go
```

```python
data = pd.read_csv("/content/innings_deliveries(1).csv")
data
```

```python
data.head()
```

```python
data.tail()
```

```python
data.isnull()
```

```python
data.dropna()
```

```python
fig = px.histogram(data['over'])
fig.show()
```

```python
fig = px.scatter(data, x='runs_batter', y='runs_total', color='team',trendline="ols",size="runs_total")
fig.show()
```

```python

batter_runs = data.groupby('batter')['runs_batter'].sum().reset_index()

fig = px.pie(batter_runs, names='batter', values='runs_batter', title='Total Runs Scored by Each Batter')
fig.show()
```

```python
# Group data by over and batter, summing runs_batter
over_batter_runs = data.groupby(['over', 'batter'])['runs_batter'].sum().reset_index()

# Create the line chart
fig = px.line(over_batter_runs, x='over', y='runs_batter', color='batter',
              title='Total Runs Scored by Each Batter Across Overs',markers=True)
fig.show()
```

```python
fig = px.bar(data, x='bowler', y='over',title='Total Runs Scored by Each Batter')
fig.show()
```

```python
batter_count = data.groupby('batter')['runs_batter'].count().reset_index()

fig = px.pie(batter_count, names='batter', values='runs_batter', title='Total Runs Scored by Each Batter')
fig.show()
```

```python
#team highest score
team_runs = data.groupby('team')['runs_total'].sum().reset_index()

fig = px.bar(team_runs, x='team', y='runs_total', title='Total Runs Scored by Each Team',color="team")
fig.show()
```

```python
#best batter who have highest run
batter_runs = data.groupby('batter')['runs_batter'].sum().reset_index()

fig = px.bar(batter_runs, x='batter', y='runs_batter',color='runs_batter', title='Total Runs Scored by Each Batter')
fig.show()
```

```python
fig = px.bar(data, x='runs_total', y='runs_total', color='team', barmode='group')
fig.show()
```

```python
#violine plot of runs
fig = px.violin(data, y='runs_batter', title='Runs Scored by Each Batter')
fig.show()
```
