```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
import plotly.express as px
import plotly.graph_objects as go
```

```python
data = pd.read_csv('/content/ElectricCarData_Clean.csv')
data
```

```python
#pie chart of power train using plotly
power_count = data['PowerTrain'].value_counts().reset_index()
power_count
```

```python
import plotly.express as px

# Assuming 'data' is your DataFrame and 'PowerTrain' is the column with power train types
power_train_counts = data['PowerTrain'].value_counts()
fig = px.pie(power_train_counts,
             values=power_train_counts.values,
             names=power_train_counts.index,
             title='Power Train Type Distribution')
fig.show()
```

# Analysis -
The pie chart shows the distribution of power train types, specifically between AWD, FWD, and RWD.
AWD: This category dominates the chart, representing approximately 39.98% . This indicates that AWD power trains are the most common among the data points.
FWD: This category comprises a significant portion  35.59% of the total. This suggests that FWD power trains are also minimum .
RWD: This category has the smallest share, representing 24.43% of the total. This indicates that RWD power trains are the least common among the data points.

```python
#bar of brand vs Euro
brand_euro = data.groupby('Brand')['PriceEuro'].mean().reset_index()
brand_euro
fig = px.bar(brand_euro, x='Brand', y='PriceEuro', title='Average Price (Euro) by Brand')
fig.show()
```

# Analysis -
The graph shows the average price (in Euros) of different brands. We can see that there's a wide range of prices across the brands.
High-Priced Brands: Some brands, such as Lightyear and Porsche, have significantly higher average prices compared to others.
Mid-Priced Brands: Many brands fall into the mid-price range, with average prices ranging from 40,000 to 80,000 Euros.
Low-Priced Brands: Some brands, like "SEAT", have the lowest average prices.



```python
px.scatter(data, x='Brand', y='FastCharge_KmH', color='PowerTrain', title='FastCharge_KmH by Brand', size='Seats')
```

```python
#line graph of brand vs topspeed_kmH
brand_topspeed = data.groupby('Brand')['TopSpeed_KmH'].mean().reset_index()
brand_topspeed
fig = px.line(brand_topspeed, x='Brand', y='TopSpeed_KmH', title='Average Top Speed by Brand',markers=True)
fig.show()
```

Analysis -
This graph shows the average top speed (in km/h) of different car brands. Each point represents a brand, and the height of the point along the vertical axis shows its top speed .Highest Top Speeds: Some brands, like Lucid, Porsche, and Tesla, reach much higher top speeds compared to others, going beyond 220 km/h. This indicates that these brands  focus on high-performance .Several brands such as Renault, Citroen, Fiat, and Smart have lower top speeds, mostly below 160 km/h.

```python
fig = go.Figure()
fig.add_trace(go.Bar(
    x=data["PowerTrain"],
    y=data["FastCharge_KmH"],
    name='FastCharge_KmH',
    marker_color='indianred'
))
fig.add_trace(go.Bar(
    x=data["PowerTrain"],
    y=data["TopSpeed_KmH"],
    name='TopSpeed_KmH',
    marker_color='lightsalmon'
))
fig.update_layout(barmode='group', xaxis_tickangle=-45)
fig.show()
```

# Analysis -
The graph shows the comparison of 2 variables: "FastCharge_kWh_50," and "Total_kWh_50" across three different time periods (years).
FastCharge_kWh: This variable shows a general upward trend, with the highest value in the last time period .
