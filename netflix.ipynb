```python
import plotly.express as px
import pandas as pd

# Calculate genre counts and select top 10
genre_counts = data['genre'].value_counts().head(10)

# Create Plotly bar chart
fig = px.bar(
    x=genre_counts.index,
    y=genre_counts.values,
    title='Top 10 Genres',
    labels={'x': 'Genre', 'y': 'Count'},
    color=genre_counts.index,
    color_discrete_sequence=px.colors.qualitative.Pastel
)

fig.show()
```

# Analysis -
This graph displays the top 10 most popular movie genres based on the number of titles available on a streaming service.

Key takeaways:

Documentary is the most popular genre: It has the highest count, with 160 titles.
Drama comes in second: It has a count of 80 titles.
Other popular genres: Comedy, Romantic Comedy, and Thriller follow closely behind.
Less popular genres: Genres like Horror, Biopic, and Action have the lowest number of titles available.
Overall, this graph tells us that Documentary and Drama are the most widely available genres on this streaming service.


```python
lang_count = data['language'].value_counts().head(10)
lang_count

fig = px.bar(
    x=lang_count.index,
    y=lang_count.values,
    title='Top 10 Languages',
    labels={'x': 'Language', 'y': 'Count'},
    color=lang_count.index,

)
fig.show()
```

This graph displays the top 10 most popular movie languages based on the number of titles available on a streaming service.
English is the most popular language: It has the highest count, with 400 titles.
Hindi comes in second: It has a count of 40 titles.
Other popular languages: Spanish, French, and Italian follow closely behind.
Less popular languages: Languages like Indonesian, Korean, and Japanese have the lowest number of titles available.
Overall, this graph tells us that English is the most widely available language on this streaming service.




```python
figure = px.bar(data[:100],
                x="year",
                y='imdb_score',
                title="Top imdb_score genre",  # Consider updating the title for clarity
                color_discrete_sequence=['blueviolet'])  # Set color
figure.show()
```

# Analysis -
This graph displays the number of movies with the top runtime for each year from 2015 to 2021.
2018 had the most movies with the top IMDb score: It has the highest count, with 120 movies.
2019 and 2020 were also strong years: They both had a count of 160 movies with the top IMDb score.
2015 and 2016 were the weakest years: They had the lowest counts, with only 10 movies each.
2017 had a moderate number of top-rated movies: It had a count of 80 movies.
Overall, this graph suggests that the quality of movies with top IMDb scores has been increasing over the years, with 2018, 2019, and 2020 being particularly strong years.
