By finding mode we have got 294 as the mode value .
we will need to use this value for dividing the sum of all the samples related to HDI , GDP and Population


```python
#lets create a new dataset by combining the necessary columns from both dataset
```

**Aggregating the data**

```python
code = df1["CODE"].unique().tolist()
country = df1["COUNTRY"].unique().tolist()
hdi = []
tc = []
td = []
sti = []
population = df1["POP"].unique().tolist()
gdp = []

for i in country:
  hdi.append((df1.loc[df1["COUNTRY"]==i,"HDI"]).sum()/294)
  tc.append((df.loc[df["location"]==i,"total_cases"]).sum())
  td.append((df.loc[df["location"]==i,"total_deaths"]).sum())
  sti.append((df1.loc[df1["COUNTRY"]==i,"STI"]).sum()/294)
  population.append((df.loc[df["location"]==i,"population"]).sum()/294)

aggregated_data = pd.DataFrame(list(zip(code,country,hdi,tc,td,sti,population)),
                               columns = ["Country Code","Country Name","HDI","TOTAL CASES","TOTAL DEATHS",
                                          "Stringency Index","Population"])
print(aggregated_data.columns)

print(aggregated_data.head())
code = aggregated_data['Country Code'].unique().tolist()
```

we have not included GDI_per_capita column because both the datasets doesn't have correct figures so we need manual data of the each coutry taken in dataset

```python
#As it is very difficult to find gdp of all the countries we'll select the samples from the dataset
```

To create subsamples we'll be selecting top 10  countries with highest number of covid 19 cases.
So we have to sort accordingly

```python
df1 = aggregated_data.sort_values(by=["TOTAL CASES"], ascending=False)
print(df1.head())
```

```python
# Top 10 Countries with Highest Covid Cases

df1 = df1.head(10)
print(df1)
```

```python
# add two more columns (GDP per capita before Covid-19, GDP per capita during Covid-19) to this dataset

df1["GDP Before Covid"] = [65279.53, 8897.49, 2100.75,
                            11497.65, 7027.61, 9946.03,
                            29564.74, 6001.40, 6424.98, 42354.41]
df1["GDP During Covid"] = [63543.58, 6796.84, 1900.71,
                            10126.72, 6126.87, 8346.70,
                            27057.16, 5090.72, 5332.77, 40284.64]
print(df1)
```

```python
px.bar(df1, x="Country Name", y="TOTAL CASES", title="Top 10 Countries with Highest Covid Cases")
```

```python
px.bar(df1,x="Country Name",y="TOTAL DEATHS",title="Top 10 Countries with Highest Covid Deaths")
```

```python
#total number of cases and total deaths in all these countries
total_cases = df1["TOTAL CASES"].sum()
total_deaths = df1["TOTAL DEATHS"].sum()
print("Total Cases:",total_cases)
print("Total Deaths:",total_deaths)

sns.barplot(x=["Total Cases","Total Deaths"],y=[total_cases,total_deaths],palette="viridis")
plt.legend(["Total Cases","Total Deaths"])
plt.title("Total Cases and Total Deaths in Top 10 Countries")
plt.show()
```

```python
fig = go.Figure()
fig.add_trace(go.Bar(
    x=df1["Country Name"],
    y=df1["TOTAL CASES"],
    name='Total Cases',
    marker_color='indianred'
))
fig.add_trace(go.Bar(
    x=df1["Country Name"],
    y=df1["TOTAL DEATHS"],
    name='Total Deaths',
    marker_color='lightsalmon'
))
fig.update_layout(barmode='group', xaxis_tickangle=-45)
fig.show()
```

```python
# Percentage of Total Cases and Deaths
total_cases = df1["TOTAL CASES"].sum()
total_deaths = df1["TOTAL DEATHS"].sum()
print("Total Cases:",total_cases)
print("Total Deaths:",total_deaths)

labels = ["Total_cases", "Total_deaths"]
values = [total_cases , total_deaths ]

fig = px.pie(df1, values=values, names=labels,
             title='Percentage of Total Cases and Deaths', hole=0.5)
fig.show()
```

```python
#Death rate of covid-19
df1["Death Rate"] = (df1["TOTAL DEATHS"].sum()/df1["TOTAL CASES"].sum())*100
print(df1)
```

```python
df1["Death Rate"] = df1["Death Rate"].astype(int)
print(df1)
```
